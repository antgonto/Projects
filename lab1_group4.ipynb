## RSA Encryption & Decryption — Integer Messages

## Objective
Implement and test RSA encryption and decryption for integer messages:
- Public key: **(e, n)**
- Private key: **(d, n)**

This is RSA implementation uses small numbers with no padding.

### Key pieces
- Choose two primes: p and q
- n = p * q -> the modulus -> the product of two primes p and q.
- e -> encryption exponent -> a number used for the public key.
- d -> decryption exponent -> a number used for the private key.

### Formulas
- Encrypt:  C = M^e mod n
- Decrypt:  M = C^d mod n

### We use of `pow(M, e, n)` instead of `M**e % n`
Computing `(M**e) % n` can be slow and memory heavy for large exponents. The use of `pow(M, e, n)`
does fast modular exponentiation efficiently.


````python
# Helper: ensure a parameter is an integer.
def _validate_int(name: str, value: int) -> None:
    if not isinstance(value, int):
        raise TypeError(f"{name} must be an int, got {type(value).__name__}")
````

### RSA implementation (integers)
We implement two functions:
   - rsa_encrypt(message, e, n)  -> ciphertext
   - rsa_decrypt(ciphertext, d, n) -> message

It assumes the message is an integer in [0, n-1].

````python
def rsa_encrypt(message: int, e: int, n: int) -> int:
    """Encrypt an integer message using RSA public key components (e, n).

    Math:
        C = M^e mod n

    Parameters:
        message (int): Plaintext integer M.
        e (int): Public exponent.
        n (int): Modulus (typically p*q).

    Returns:
        int: Ciphertext integer C.
    """
    # --- Type checks ---
    _validate_int("message", message)
    _validate_int("e", e)
    _validate_int("n", n)

    # --- Basic value checks ---
    # n must be > 1 for modular arithmetic to make sense.
    if n <= 1:
        raise ValueError("n must be > 1")

    # Exponents must be positive integers.
    if e < 1:
        raise ValueError("e must be >= 1")

     # RSA works modulo n, so we keep messages inside [0, n-1].
    if not (0 <= message < n):
        raise ValueError(f"message must satisfy 0 <= message < n (got message={message}, n={n})")

    # --- Core RSA operation ---
    # pow(message, e, n) efficiently computes (message**e) % n
    return pow(message, e, n)
````

````python
def rsa_decrypt(ciphertext: int, d: int, n: int) -> int:
    """Decrypt an integer ciphertext using RSA private key components (d, n).

    Math:
        M = C^d mod n

    Parameters:
        ciphertext (int): Ciphertext integer C.
        d (int): Private exponent.
        n (int): Modulus (same n used in encryption).

    Returns:
        int: Decrypted plaintext integer M.
    """
    # --- Type checks ---
    _validate_int("ciphertext", ciphertext)
    _validate_int("d", d)
    _validate_int("n", n)

    # --- Basic value checks ---
    if n <= 1:
        raise ValueError("n must be > 1")
    if d < 1:
        raise ValueError("d must be >= 1")

    # Ciphertext values are also elements of Z_n (0..n-1).
    if not (0 <= ciphertext < n):
        raise ValueError(f"ciphertext must satisfy 0 <= ciphertext < n (got ciphertext={ciphertext}, n={n})")

    # --- Core RSA operation ---
    return pow(ciphertext, d, n)
````

## Test Vectors

A test vector is a known input/output pair used to verify correctness.
For each set we are given:
- \(n\), \(e\), \(d\)
- a plaintext message \(M\)
- the expected ciphertext \(C\)

If our `rsa_encrypt` function is correct, it should produce the expected ciphertext.
If our `rsa_decrypt` function is correct, decrypting that ciphertext should recover the original message.

````python
# Each dictionary below describes one RSA key set + a plaintext message.
# We keep p and q only for reference because our functions use (e, n) and (d, n).
test_sets = [
    {"set": "A", "p": 3,  "q": 11, "n": 33, "e": 3, "d": 7,  "M": 15, "ExpectedCiphertext": 9},
    {"set": "B", "p": 5,  "q": 11, "n": 55, "e": 3, "d": 27, "M": 42, "ExpectedCiphertext": 3},
    {"set": "C", "p": 7,  "q": 13, "n": 91, "e": 5, "d": 29, "M": 10, "ExpectedCiphertext": 82},
]
````

## Test Runner (Encrypt → Decrypt → Assert)

We will loop through each test set and perform these steps:
1. Read \(M\), \(e\), \(d\), \(n\), and the expected ciphertext.
2. Compute `C = rsa_encrypt(M, e, n)`.
3. Compute `M2 = rsa_decrypt(C, d, n)`.
4. Print results so we can see what happened.
5. Use `assert` to automatically check:
   - `C` equals the expected ciphertext
   - `M2` equals the original plaintext `M`

If an `assert` fails, Python will raise an error and tell us which test failed.


````python
def tests():
    for t in test_sets:
        name = t["set"]
        n, e, d = t["n"], t["e"], t["d"]
        M = t["M"]
        expected_c = t["ExpectedCiphertext"]

        C = rsa_encrypt(M, e, n)
        M2 = rsa_decrypt(C, d, n)

        print(f"Set {name}:")
        print(f"  Public key  (e, n) = ({e}, {n})")
        print(f"  Private key (d, n) = ({d}, {n})")
        print(f"  Plaintext M        = {M}")
        print(f"  Ciphertext C       = {C}")
        print(f"  Decrypted M2       = {M2}")
        print()

        assert C == expected_c, f"Set {name}: expected C={expected_c}, got C={C}"
        assert M2 == M,         f"Set {name}: expected M2={M}, got M2={M2}"

    print("✅ All RSA tests passed (ciphertext and round-trip decryption).")
````

````python
tests()
````

# **Hashing**

A hash function in cryptography is a mathematical algorithm that takes an input (called a message or data) of any size and produces a fixed-size output, typically a string of hexadecimal characters known as a hash value or digest. Key properties include:

•	Determinism: The same input always produces the same output.

•	Efficiency: It's quick to compute the hash for any input.

•	One-way: It's computationally infeasible to reverse the hash to get the original input (preimage resistance).

•	Avalanche effect: A small change in input produces a significantly different output.

•	Collision resistance: It's hard to find two different inputs with the same hash. Cryptographic hash functions like SHA-256 are used for data integrity, digital signatures, password storage, and more.

We implemented the following Python function to compute the SHA-256 hash of the message:

m = **"Information Systems Security Engineering and Management Spring 2026"**



````python
import hashlib  # Import the hashlib library for cryptographic hashing

message = "Information Systems Security Engineering and Management Spring 2026"  # The message to hash

# Create a SHA-256 hash object and update it with the encoded message
hash_object = hashlib.sha256(message.encode('utf-8'))

# Get the hexadecimal digest (hash value)
hex_dig = hash_object.hexdigest()

print("Hash value:", hex_dig)
````

Running this outputs:

**Hash Verification**

We implemented the following Python function for hash verification:

````python
import hashlib  # Import hashlib for hashing

def hash_verification(data, original_hash):
    """
    Verifies the integrity of the data by computing its SHA-256 hash
    and comparing it to the provided original_hash.

    Args:
    data (str): The data to verify.
    original_hash (str): The expected hash value.

    Returns:
    bool: True if hashes match (data intact), False otherwise.
    """
    # Compute the SHA-256 hash of the data
    hash_object = hashlib.sha256(data.encode('utf-8'))
    new_hash = hash_object.hexdigest()

    # Compare the new hash to the original
    return new_hash == original_hash
````

Now, using the function to verify the integrity of the original message:

````python
# Original message and its noted hash
data = "Information Systems Security Engineering and Management Spring 2026"
original_hash = "01e6fcce29e587bb0b1e9e171d17eb68caca4c23b66b84414815ed0ceac70d6b"

# Verify
is_intact = hash_verification(data, original_hash)
print("Data integrity verified:", is_intact)
````

Running this outputs:

Data integrity verified: True

To demonstrate corruption, we modify the data slightly by changing the year to 2027:

````python
# Modified (corrupted) data
corrupted_data = "Information Systems Security Engineering and Management Spring 2027"  # Changed year

# Verify against original hash
is_intact_corrupted = hash_verification(corrupted_data, original_hash)
print("Corrupted data integrity verified:", is_intact_corrupted)
````

Running this outputs:

Corrupted data integrity verified: False

This confirms the hash detects changes and our hash verification function detects any changes to the original message.

# Code Components Overview

This notebook does not define any Python **classes** (i.e., no `class` keyword is used). Instead, it is organized around a set of **functions**, each with a clearly scoped responsibility. Below is a description of every function defined in this notebook.

---

## 1. `_validate_int(name, value)`

**Location:** RSA section — helper function

**Purpose:** A private utility that checks whether a given value is an `int`. If the value is not an integer, it raises a `TypeError` with a descriptive message that includes the parameter name and the actual type received.

**Parameters:**
- `name` (`str`): The human-readable name of the parameter being validated (used in the error message).
- `value` (`int`): The value to validate.

**Returns:** `None` — it either passes silently or raises `TypeError`.

**Why it exists:** RSA arithmetic requires integer operands. Catching bad types early produces clear, actionable error messages instead of confusing math errors deeper in the call stack.

---

## 2. `rsa_encrypt(message, e, n)`

**Location:** RSA section

**Purpose:** Encrypts an integer plaintext `message` using the RSA public key `(e, n)`. It implements the core RSA encryption formula:

```
C = M^e mod n
```

**Parameters:**
- `message` (`int`): The plaintext integer `M`; must satisfy `0 <= message < n`.
- `e` (`int`): The public exponent; must be `>= 1`.
- `n` (`int`): The RSA modulus (product of two primes `p * q`); must be `> 1`.

**Returns:** `int` — the ciphertext `C`.

**Key implementation detail:** Uses Python's built-in `pow(message, e, n)` which performs fast modular exponentiation, avoiding the memory overhead of computing `message**e` before taking the modulus.

---

## 3. `rsa_decrypt(ciphertext, d, n)`

**Location:** RSA section

**Purpose:** Decrypts an RSA ciphertext back to the original integer plaintext using the RSA private key `(d, n)`. It implements the core RSA decryption formula:

```
M = C^d mod n
```

**Parameters:**
- `ciphertext` (`int`): The ciphertext integer `C`; must satisfy `0 <= ciphertext < n`.
- `d` (`int`): The private exponent; must be `>= 1`.
- `n` (`int`): The RSA modulus — the same value used during encryption; must be `> 1`.

**Returns:** `int` — the recovered plaintext `M`.

**Key implementation detail:** Mirrors `rsa_encrypt` exactly, relying on the mathematical property of RSA that applying the private exponent to the ciphertext reverses the public-exponent transformation.

---

## 4. `tests()`

**Location:** RSA section — test runner

**Purpose:** Iterates over the predefined `test_sets` list (test vectors with known inputs and expected outputs) to verify that `rsa_encrypt` and `rsa_decrypt` are correct.

**Parameters:** None — uses the module-level `test_sets` list directly.

**Returns:** `None` — prints results and raises `AssertionError` if any check fails.

**What it checks for each test vector:**
1. `rsa_encrypt(M, e, n)` produces the expected ciphertext `C`.
2. `rsa_decrypt(C, d, n)` recovers the original plaintext `M` (round-trip correctness).

---

## 5. `hash_verification(data, original_hash)`

**Location:** Hashing section

**Purpose:** Verifies the integrity of a piece of data by re-computing its SHA-256 hash and comparing it to a previously recorded hash value.

**Parameters:**
- `data` (`str`): The string whose integrity is being checked.
- `original_hash` (`str`): The expected SHA-256 hex digest.

**Returns:** `bool` — `True` if the newly computed hash matches `original_hash` (data is unchanged), `False` otherwise (data has been tampered with or corrupted).

**Key implementation detail:** Uses `hashlib.sha256` from the Python standard library and encodes the input string as UTF-8 bytes before hashing, which is the standard approach for hashing text data.
